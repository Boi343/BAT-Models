```python
import pandas as pd
import numpy as np
import requests
import json
import datetime as dt
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
def getTradingData(fromDate, toDate, ticker):
    """
    Khởi tạo chuỗi thời gian giao dịch thực tế.
    
    Args:
        **fromDate** (datetime): Ngày bắt đầu \n
        **toDate** (datetime): Ngày kết thúc
        **ticker** (string): Tên ticker
    
    Returns:
        pandas.DataFrame.
    """
    
    fDate = int(fromDate.timestamp())
    tDate = int(toDate.timestamp())
    data = requests.get(f"https://api.vietstock.vn/ta/history?symbol={ticker}&resolution=D&from={fDate}&to={tDate}").json()
    data2 = json.loads(data)
    df0 = pd.DataFrame(data2)
#     df0["t"] = pd.to_datetime(df0["t"],unit="s").apply(lambda x: dt.datetime(x.year,x.month,x.day))
    df0["t"] = pd.to_datetime(df0["t"],unit="s")
    df0["symbol"]=ticker
    df0.columns=['Date','Open','High','Low','Close','Volume','Status','Ticker']
    return df0
```

```python
def convertToModel(rawData):

    ###################################################################################.
    
    pvtData = data.pivot_table(index="Date",columns="Ticker",values=["Close", "Open","High","Low","Volume"], 
                               aggfunc="first").iloc[::-1].stack("Ticker")
    port = pvtData.fillna("")
    ##################################################################################.
    port.unstack()

    port ["C"] = port["Close"]
    port ["O"] = port["Open"]
    port ["H"] = port["High"]
    port ["L"] = port["Low"]


    for i in range (1,9):
        port["C(-"+str(i)+")"]=port.groupby("Ticker")["C"].shift(-i)
        port["O(-"+str(i)+")"]=port.groupby("Ticker")["O"].shift(-i)
        port["H(-"+str(i)+")"]=port.groupby("Ticker")["H"].shift(-i)
        port["L(-"+str(i)+")"]=port.groupby("Ticker")["L"].shift(-i)

    ##################################################################################.
    CandleType=port.copy()

    CandleType["T(-1)_Diff"]=CandleType["C"]>CandleType["C(-1)"]
    CandleType["Body"]=CandleType["C"] - CandleType["O"]
    CandleType["Body(T-1)"]=CandleType["C(-1)"] - CandleType["O(-1)"]


    G_H=(CandleType["T(-1)_Diff"]>=0) & (CandleType["Body"]>0)
    G_S=(CandleType["T(-1)_Diff"]>=0) & (CandleType["Body"]<0)
    R_H=(CandleType["T(-1)_Diff"]<0) & (CandleType["Body"]>0)
    R_S=(CandleType["T(-1)_Diff"]<0) & (CandleType["Body"]<0)
    Doji=(CandleType["Body"]==0)

    CandleType["Candle_Type"]=np.where(Doji, "Doji",np.where(G_H,"Green-Hollow",
                                                       np.where(G_S,"Solid Green",np.where(R_H,"Red-Hollow","Solid Red"))))
    # Basic specifications

    Model=CandleType.copy()

    # Trend
    Model.loc [Model["C"]<Model["C(-1)"], "Diff_1_D"]="Down"
    Model.loc [Model["C"]>Model["C(-1)"], "Diff_1_U"]="Up"

    for i in range (1,8):
        Model.loc [Model["C(-"+ str(i) + ")"]<Model["C(-"+ str(i+1)+")"], "Diff_" + str(i+1)+"_D" ]="Down"
        Model.loc [Model["C(-"+ str(i) + ")"]>Model["C(-"+ str(i+1)+")"], "Diff_" + str(i+1)+"_U" ]="Up"
        Model["Body(-"+ str(i) + ")"]=abs(Model.groupby("Ticker")["Body"].shift(-i))
        
        
    for i in range (1,6):
        Model.loc [((Model["C(-"+ str(i) + ")"]==Model["C(-"+ str(i+1)+")"]) | (Model["C(-"+ str(i+1) + ")"]==Model["C(-"+ str(i+2)+")"]))|
                   (((Model["H(-"+ str(i+2) + ")"]>Model["H(-"+ str(i+1)+")"]) & (Model["L(-"+ str(i+2) + ")"]<Model["L(-"+ str(i+1)+")"])) &
                    ((Model["H(-"+ str(i+1) + ")"]>Model["H(-"+ str(i)+")"]) & (Model["L(-"+ str(i+1) + ")"]<Model["L(-"+ str(i)+")"]))), "Sideway_"+str(i)+"-Candle"]="Sideway"

    
    
    
    
    
#     Model.loc [((Model["C(-1)"]==Model["C(-2)"]) | (Model["C(-2)"]==Model["C(-3)"]))|
#                (((Model["H(-3)"]> Model["H(-2)"]) & (Model["L(-3)"]< Model["L(-2)"])) &
#                 ((Model["H(-2)"]> Model["H(-1)"]) & (Model["L(-2)"]< Model["L(-1)"]))), "Sideway"]="Sideway"


#     Model.loc [((Model["C(-2)"]==Model["C(-3)"]) | (Model["C(-3)"]==Model["C(-4)"]))|
#                (((Model["H(-3)"]> Model["H(-2)"]) & (Model["L(-3)"]< Model["L(-2)"])) &
#                 ((Model["H(-2)"]> Model["H(-1)"]) & (Model["L(-2)"]< Model["L(-1)"]))), "Sideway"]="Sideway"
    
    
    
#     Model.loc [(((Model["C(-2)"]==Model["C(-3)"])| (Model["C(-3)"]==Model["C(-4)"]))|
#                (((Model["H(-4)"]> Model["H(-3)"]) & (Model["L(-4)"]< Model["L(-3)"])) &
#                 ((Model["H(-3)"]> Model["H(-2)"]) & (Model["L(-3)"]< Model["L(-2)"]))), "Sideway_2 candles"]="Sideway"
    
    
    
#     Model.loc [((Model["C(-3)"]==Model["C(-4)"]) | (Model["C(-4)"]==Model["C(-5)"]))|
#                (((Model["H(-5)"]> Model["H(-4)"]) & (Model["L(-5)"]< Model["L(-4)"])) &
#                 ((Model["H(-4)"]> Model["H(-3)"]) & (Model["L(-4)"]< Model["L(-3)"]))), "Sideway_3 candles"]="Sideway"
    
    
#     Model.loc [((Model["C(-6)"]==Model["C(-5)"]) | (Model["C(-5)"]==Model["C(-4)"]))|
#                (((Model["H(-6)"]> Model["H(-5)"]) & (Model["L(-6)"]< Model["L(-5)"])) &
#                 ((Model["H(-5)"]> Model["H(-4)"]) & (Model["L(-5)"]< Model["L(-4)"]))), "Sideway_4 candles"]="Sideway"
    
    
#     Model.loc [((Model["C(-7)"]==Model["C(-6)"]) | (Model["C(-6)"]==Model["C(-5)"]))|
#                (((Model["H(-7)"]> Model["H(-6)"]) & (Model["L(-7)"]< Model["L(-6)"])) &
#                 ((Model["H(-6)"]> Model["H(-5)"]) & (Model["L(-6)"]< Model["L(-5)"]))), "Sideway_5 candles"]="Sideway"
    
    
#     Model.loc [(Model["C(-1)"]==Model["C(-2)"]) | (Model["C(-2)"]==Model["C(-3)"])| (Model["C(-3)"]==Model["C(-4)"]), "Sideway_2 candles"]="Sideway"
#     Model.loc [(Model["C(-3)"]==Model["C(-4)"]) | (Model["C(-4)"]==Model["C(-5)"]), "Sideway_3 candles"]="Sideway"
#     Model.loc [(Model["C(-5)"]==Model["C(-6)"]) | (Model["C(-6)"]==Model["C(-7)"]), "Sideway_4 candles"]="Sideway"

    # Define Shadow Up, Shadow Down, Body

    Model.loc[Model["C"] >= Model["O"], "shadowUP"] = Model["H"] - Model["C"]
    Model.loc[Model["C"] <  Model["O"], "shadowUP"] = Model["H"] - Model["O"]
    Model["shadowLOW"] = Model["H"] - Model["L"] - Model["shadowUP"] - abs(Model["Body"]) 
   


    for i in range (1,3):
        Model["shadowUP(-"+ str(i) + ")"]=abs(Model.groupby("Ticker")["shadowUP"].shift(-i))
        Model["shadowLOW(-"+ str(i) + ")"]=abs(Model.groupby("Ticker")["shadowLOW"].shift(-i))

    # Conditon of 3-4 Candles Pattern
    for i in range (1,3):
        Model["H(-"+str(i)+")"]=Model.groupby("Ticker")["H"].shift(-i)
        Model["L(-"+str(i)+")"]=Model.groupby("Ticker")["L"].shift(-i)

        
    Model["whiteBody"]=Model["C"]>Model["O"]
    Model["blackBody"]=Model["O"]>Model["C"]
    Model["largeBodyMinimum"] = 0.02
    Model["smallBodyMaximum"]=0.0025
    Model.loc[(((Model["C"]>=(Model["O"]*(1+Model["largeBodyMinimum"]))) &
             (Model["whiteBody"]==True))|
             ((Model["C"]<=(Model["O"]*(1-Model["largeBodyMinimum"]))) &
             (Model["blackBody"]==True))),"largeBody"]=True

   
  #############################################################.
    Model["identicalBodies"]=abs(abs(Model["O(-1)"]-Model["C(-1)"])-abs(Model["O"]-Model["C"])) <= (abs(Model["O"]-Model["C"])*0.3)
    Model["identicalBodies(-1)"]= Model.groupby("Ticker")["identicalBodies"].shift(-1)
    
    
    
    Model.loc [ (((Model ["whiteBody"] == True) & 
                  (Model ["H"] >= (Model["C"]*(1+ Model["largeBodyMinimum"]))))|
                  ((Model ["blackBody"] == True) & 
                  (Model ["H"] >= (Model["O"]*(1+ Model["largeBodyMinimum"]))))),
                  "largeUpperShadow"]=True

    Model["largeUpperShadow(-1)"]= Model.groupby("Ticker")["largeUpperShadow"].shift(-1)
    Model.loc [ (((Model ["O"] >= (Model["C"]*(1- Model["smallBodyMaximum"]))) &
                (Model ["whiteBody"] == True))|
              ((Model ["C"] >= (Model["O"]*(1- Model["smallBodyMaximum"]))) &
                  (Model ["blackBody"] == True))),
             "smallBody"]=True

     ########################################################################################.
    for i in range (1,8):
        Model["blackBody(-" + str(i) +")"]=Model.groupby("Ticker")["blackBody"].shift(-i)
        Model["whiteBody(-" + str(i) +")"]=Model.groupby("Ticker")["whiteBody"].shift(-i)
        Model["largeBody(-" + str(i) +")"]=Model.groupby("Ticker")["largeBody"].shift(-i)
        Model["smallBody(-" + str(i) +")"]=Model.groupby("Ticker")["smallBody"].shift(-i)

    #/*GAPS*/################################################################
   
    # upGap
    Model["upGap_condition_1"]= np.where((Model["blackBody(-1)"]==True) & (Model["whiteBody"]==True) &
                                           (Model["O"]>Model["O(-1)"]),"TRUE","")

    Model["upGap_condition_2"]= np.where((Model["blackBody(-1)"]==True) & (Model["blackBody"]==True) &
                                           (Model["C"]>Model["O(-1)"]),"TRUE","")

    Model["upGap_condition_3"]= np.where((Model["whiteBody(-1)"]==True) & (Model["whiteBody"]==True) &
                                           (Model["O"]>Model["C(-1)"]),"TRUE","")

    Model["upGap_condition_4"]= np.where((Model["whiteBody(-1)"]==True) & (Model["blackBody"]==True) &
                                           (Model["C"]>Model["C(-1)"]),"TRUE","")


    Model.loc[((Model["upGap_condition_1"]=="TRUE")|
                          (Model["upGap_condition_2"]=="TRUE")|
                          (Model["upGap_condition_3"]=="TRUE")|
                          (Model["upGap_condition_4"]=="TRUE")
                ),"upGap"]=True

    for i in range (1,3):
        Model["upGap(-" + str(i) +")"]=Model.groupby("Ticker")["upGap"].shift(-i)

    Model["upGap"].unstack().fillna("")
    #-----------------------#
    # downGap

    Model["downGap_condition_1"]= np.where((Model["blackBody(-1)"]==True) & (Model["whiteBody"]==True) &
                                           (Model["C"]<Model["C(-1)"]),"TRUE","")

    Model["downGap_condition_2"]= np.where((Model["blackBody(-1)"]==True) & (Model["blackBody"]==True) &
                                           (Model["O"]<Model["C(-1)"]),"TRUE","")

    Model["downGap_condition_3"]= np.where((Model["whiteBody(-1)"]==True) & (Model["whiteBody"]==True) &
                                           (Model["C"]<Model["O(-1)"]),"TRUE","")

    Model["downGap_condition_4"]= np.where((Model["whiteBody(-1)"]==True) & (Model["blackBody"]==True) &
                                           (Model["O"]<Model["O(-1)"]),"TRUE","")

    Model.loc[((Model["downGap_condition_1"]=="TRUE")|
                          (Model["downGap_condition_2"]=="TRUE")|
                          (Model["downGap_condition_3"]=="TRUE")|
                          (Model["downGap_condition_4"]=="TRUE")
                ),"downGap"]=True

    for i in range (1,3):
        Model["downGap(-" + str(i) +")"]=Model.groupby("Ticker")["downGap"].shift(-i)
 
    
    
  
        
        
    return Model
```

```python
import plotly
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
from datetime import datetime
import datetime

########Candle Stickss Combination #####################33.
signal_BUY = Model[Model.iloc[:,:].isin(["BUY"])].dropna(how="all").assign(signal= "BUY")
signal_SELL = Model[Model.iloc[:,:].isin(["SELL"])].dropna(how="all").assign(signal= "SELL")
signal_all = pd.concat([signal_BUY,signal_SELL])


def merge(row):
    data = []
    for column in list(signal_all.columns[:-1]):        
        if row[column] == "BUY":
            data.append(column)
    for column in list(signal_all.columns[:-1]):        
        if row[column] == "SELL":
            data.append(column)           
    return data


signal_all["Candle Types"]  = signal_all.apply(merge, axis=1)
signal_all


signal_CANDLE = signal_all.pivot_table(index="Date", columns="Ticker", values=["signal","Candle Types"],aggfunc="first").iloc[::-1].stack("Ticker").reindex(columns=["signal","Candle Types"])
signal_CANDLE.unstack().fillna("")

######################
Data_Btest = pd.concat([Model, signal_CANDLE], axis=1).fillna("")
##################


def sample_plot(model, days):
    plot = plotly.tools.make_subplots(rows=2, cols=5, print_grid=False)
#     plotly.tools.set_credentials_file(username='tranquocboi', api_key='ytE9Aem5qg3lvRcPnBxy')
   
    
    plotly.offline.init_notebook_mode(connected=True)
    for i in range (1,11): 
        
        random_candle = Data_Btest[Data_Btest[model].isin(["BUY", "SELL"])].sample(1).iloc[0]
        sample_date = random_candle.name[0]
        sample_symbol = random_candle.name[1]



        final=Data_Btest.loc[pd.IndexSlice[:,sample_symbol],:]
  
        
        # Temp dataframe to calculate row range number
        temp=final.reset_index(level=1)
        t=temp.index.get_loc(sample_date)


        # Pre-Data for backtest
        sample_period=final.iloc[(t-days):(t+1),:]
        sample_period

        # Data for backtest
        test=sample_period.reset_index(level=1)
        test


       
        trace = go.Candlestick(x=test.index,
                    open=test['Open'],
                    high=test['High'],
                    low=test['Low'],
                    close=test['Close'],
                    name  = sample_symbol)


        if i<=5:
                        plot.append_trace(trace, 1, i)
        else:
                        plot.append_trace(trace, 2, i-5)

        format_xaxis = dict(tickformat= '%d/%m/%y', tickangle=90, rangeslider = dict(visible=False,autorange=True),type="date")      
        format_yaxis = dict(tickformat= ".0f")        
        all_xaxis = {f"xaxis{i}": format_xaxis for i in range(1,11)}
        all_yaxis = {f"yaxis{i}": format_yaxis for i in range(1,11)}

        plot['layout'].update(height=500, width=1000, plot_bgcolor="#000000", font=dict(size=9),
                                          clickmode="select", hovermode="closest",
                                          **all_xaxis, **all_yaxis)

   
    return py.iplot(go.Figure(data=plot))
```
